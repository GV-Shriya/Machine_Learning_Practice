We now have a non-null dataframe with no outliers. But the object-type categorical data needs to be converted to some numeric-valued data to understand and analyze the dataset and perform Linear Regresion.

```python
cat_var_list =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

def binary_map(x):
    return x.map({'yes': 1, "no": 0})

df[cat_var_list] = df[cat_var_list].apply(binary_map)

df
```

The categorical data has been converted to numeric data. Next we need to convert 'furnishingstatus' values to some type of numeric values as well. In the above conversion, the columns were 2-level categorical variables so it was easy to convert with yes corresponding to 1 and no corresponding to 0. But furnishingstatus has 3 levels. When you have a categorical variable with n-levels, the idea is to create a dummy variable by building â€˜n-1â€™ variables, indicating the levels. Pandas has get_dummies method for it.

```python
furnishing_status = pd.get_dummies(df['furnishingstatus'])

furnishing_status
```

We donâ€™t need three columns. We can drop the furnished column, as it can be identified with just the last two columns values where:
00 - furnished ,
01 - unfurnished ,
10 - semi-furnished

```python
# Dropping the first column from furnishing_status dataset
furnishing_status = pd.get_dummies(df['furnishingstatus'], drop_first = True)

# Adding the converted finishing status to the original df dataframe
df = pd.concat([df, furnishing_status], axis = 1)

# Dropping 'furnishingstatus' as we have created the dummies for it
df.drop(['furnishingstatus'], axis = 1, inplace = True)

df
```

```python
X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'semi-furnished','unfurnished']]
y = df.price
```

X contains all the columns except price and y contains only the price column

```python
X
```

```python
y
```

Splitting the data into training and testing datasets

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

Building the model

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

Predicting using testing dataset

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test 
```

```python
r2_score(y_test, y_pred)
```

The accuracy of my model is 76%
